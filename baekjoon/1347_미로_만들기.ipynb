```python
n = int(input())
a = input()
d = [[1, 0], [0, -1], [-1, 0], [0, 1]]
b = [[0, 0]]
direction = 0
real = [0, 0]
for i in a:
    if i == 'R':
        direction = direction + 1 if direction != 3 else 0
    if i == 'L':
        direction = direction - 1 if direction != 0 else 3
    if i == 'F':    # 방향에 맞게 전진하기
        real = [real[0] + d[direction][0], real[1] + d[direction][1]]
        b.append(real)    # 내가 갔던 칸을 기록해두기
x_min, x_max, y_min, y_max = 100, 0, 100, 0   # 총 이동 반경 보기 위해
for i in b:
    x_min = min(x_min, i[0])
    x_max = max(x_max, i[0])
    y_min = min(y_min, i[1])
    y_max = max(y_max, i[1])
n = y_max - y_min + 1    # 총 x 축 반경
m = x_max - x_min + 1    # 총 y 축 반경
miro = [['#'] * n for _ in range(m)]
for i in range(len(b)):    # 기초 점을 (0, 0)으로 초기화 하는 과정
    b[i][0] -= x_min
    b[i][1] -= y_min
for x, y in b:    # 방문 칸을 '.' 으로 만들기
    miro[x][y] = '.'
for i in miro:
    print(*i, sep='')
```
