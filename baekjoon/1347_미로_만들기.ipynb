```python
n = int(input())
a = input()
d = [[1, 0], [0, -1], [-1, 0], [0, 1]]
b = [[0, 0]]
direction = 0
real = [0, 0]
for i in a:
    if i == 'R':
        direction = direction + 1 if direction != 3 else 0
    if i == 'L':
        direction = direction - 1 if direction != 0 else 3
    if i == 'F':    #   
        real = [real[0] + d[direction][0], real[1] + d[direction][1]]
        b.append(real)    #    
x_min, x_max, y_min, y_max = 100, 0, 100, 0   #     
for i in b:
    x_min = min(x_min, i[0])
    x_max = max(x_max, i[0])
    y_min = min(y_min, i[1])
    y_max = max(y_max, i[1])
n = y_max - y_min + 1    #  x  
m = x_max - x_min + 1    #  y  
miro = [['#'] * n for _ in range(m)]
for i in range(len(b)):    #   (0, 0)   
    b[i][0] -= x_min
    b[i][1] -= y_min
for x, y in b:    #   '.'  
    miro[x][y] = '.'
for i in miro:
    print(*i, sep='')
```
