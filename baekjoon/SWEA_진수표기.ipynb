```python
for tc in range(int(input())):
    n = int(input())
    h = input()
    s = ''.join([bin(int(i, 16))[2:].zfill(4) for i in h])
    s = [int(s[i:i + 7], 2) for i in range(0, 7 * (len(s) // 7), 7)]
    print(f'#{tc + 1}', end=' ')
    print(*s)
```

---

`int(i, 16)` : i 는 16진수인데 10진수로 바꿔주세요<br>
`bin(~)[2:]` : ~ 를 이진수로 바꿔주시는데 맨 앞 2개는 부호관련이니 빼주세요<br>
`''.join()` : 하나의 문자열로 보고 이따가 7개씩 자를게요<br>
---
`int(s[i:i + 7], 2)` : s[7마디] 는 2진수인데 10진수로 바꿔주세요<br>
`for i in range(0, 7 * (len(s) // 7), 7)` : 7마디씩 반복할 건데 `7 * (len(s) // 7)` 처럼 7의 배수만 생각하고 나머지는 무시할 거에요<br>
