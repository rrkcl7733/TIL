```python
from collections import deque


def bfs(n):
    global cnt, minV
    visited[n] = 1
    q = deque([(n, 1)])
    while q:
        x, cost = q.popleft()
        visited[x] = 1
        hubo = [2 * x, x + 1, x - 1]
        for i in hubo:
            if i == k:
                if cost < minV:
                    minV = cost
                elif cost == minV:
                    cnt += 1

            elif 0 <= i < big and not visited[i]:
                q.append((i, cost + 1))


n, k = map(int, input().split())
cnt = 1
big = min(2 * max(n, k) + 1, 100001)
visited = [0] * big
minV = 10e9
if n == k:
    print(0, 1, sep='\n')
    exit()
bfs(n)
print(minV, cnt, sep='\n')
```
