```python
import sys
input = sys.stdin.readline

n = int(input())
a = [list(map(int, input().split())) for _ in range(n)]
for i in range(1, n):
    for j in range(3):
        if not j:
            a[i][j] += min(a[i - 1][j + 1], a[i - 1][j + 2])
        elif j == 1:
            a[i][j] += min(a[i - 1][j - 1], a[i - 1][j + 1])
        else:
            a[i][j] += min(a[i - 1][j - 1], a[i - 1][j - 2])
print(min(a[n - 1]))
```
