```python
p = [[-1, 0], [0, 1], [1, 0], [0, -1]]
n, m = map(int, input().split())
r, c, d = map(int, input().split())
arr = [list(map(int, input().split())) for _ in range(n)]
arr[r][c] = -1
while 1:
    if arr[r - 1][c] and arr[r][c - 1] and arr[r + 1][c] and arr[r][c + 1]:
        nr, nc = r - p[d][0], c - p[d][1]
        if arr[nr][nc] == 1:
            break
        r, c = nr, nc
        continue
    d = d - 1 if d else 3
    nr, nc = r + p[d][0], c + p[d][1]
    if not arr[nr][nc]:
        arr[nr][nc] = -1
        r, c = nr, nc
ans = 0
for i in arr:
    ans += i.count(-1)
print(ans)
```
