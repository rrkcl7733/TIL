```python
def turn(t):
    global dice
    a, b, c, d, e, f = dice[0], dice[1], dice[2], dice[3], dice[4], dice[5]
    if t == 1:
        dice = [d, b, a, f, e, c]
    elif t == 2:
        dice = [c, b, f, a, e, d]
    elif t == 3:
        dice = [e, a, c, d, f, b]
    else:
        dice = [b, f, c, d, a, e]


d = [0, [0, 1], [0, -1], [-1, 0], [1, 0]]
n, m, x, y, k = map(int, input().split())
arr = [list(map(int, input().split())) for _ in range(n)]
order = list(map(int, input().split()))
dice = [0] * 6
for i in order:
    nx, ny = x + d[i][0], y + d[i][1]
    if 0 <= nx < n and 0 <= ny < m:
        x, y = nx, ny
        turn(i)
        if not arr[x][y]:
            arr[x][y] = dice[-1]
        else:
            dice[-1] = arr[x][y]
            arr[x][y] = 0
        print(dice[0])
```
