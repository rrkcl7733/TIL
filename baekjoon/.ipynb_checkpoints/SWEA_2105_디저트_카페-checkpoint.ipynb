```python
d = [[1, 1], [1, -1], [-1, -1], [-1, 1]]  # ↘  ↙  ↖  ↗


def dfs(i, j, move, cnt):
    global ans

    # 직진 or 꺾기
    temp = move + 2 if move < 3 else move + 1

    for k in range(move, temp):
        ni, nj = i + d[k][0], j + d[k][1]

        if start == (ni, nj):       # 다시 원위치인 상황
            ans = max(ans, cnt)
            return

        if 0 <= ni < n and 0 <= nj < n and conquer[arr[ni][nj]] == 0:
            conquer[arr[ni][nj]] = 1
            dfs(ni, nj, k, cnt + 1)
            conquer[arr[ni][nj]] = 0


for tc in range(int(input())):
    n = int(input())
    arr = [list(map(int, input().split())) for _ in range(n)]

    conquer, ans = [0] * 101, -1

    for i in range(n):
        for j in range(n):
            start = (i, j)
            conquer[arr[i][j]] = 1
            dfs(i, j, 0, 1)
            conquer[arr[i][j]] = 0
    print(f'#{tc + 1} {ans}')
```
