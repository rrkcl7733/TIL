[](https://www.acmicpc.net/problem/1107)

```python
target = int(input())
ans = abs(100 - target) # ++ or --    -> 
M = int(input())
if M: #     
    broken = set(input().split())
else:
    broken = set()

#    500,000   
#      1,000,000  
for num in range(1000001):
    for N in str(num):
        if N in broken: #         
            break
    else: #      
            # min(,    +    )
        ans = min(ans, len(str(num)) + abs(num - target))

print(ans)
```
