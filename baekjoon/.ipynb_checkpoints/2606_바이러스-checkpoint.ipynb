```python
import sys
from collections import deque

def bfs(i):
    global cnt
    visited[i] = 1
    q.append(i)
    while q:
        y = q.popleft()
        for x in range(comp + 1):
            if arr[y][x] and visited[x] == 0:
                q.append(x)
                visited[x] = 1
                cnt += 1


comp = int(sys.stdin.readline())
n = int(sys.stdin.readline())
temp = [list(map(int, sys.stdin.readline().split())) for _ in range(n)]
visited = [0] * (comp + 1)
q = deque()
arr = [[0] * (comp + 1) for _ in range(comp + 1)]
for i in temp:
    arr[i[0]][i[1]], arr[i[1]][i[0]] = 1, 1
cnt = 0
bfs(1)
print(cnt)
```

---

간선이 양방향이기 때문에 `arr[i[0]][i[1]], arr[i[1]][i[0]] = 1, 1` 으로 양측 탐색을 하려했음
처음엔 cnt = 1로 설정했고 1번 컴퓨터부터 조사하나 아무것도 연결되지 않았을 때 0을 출력하지 않아 틀림.
<br><br>
애초에 cnt = 0, bfs(1)로 설정해두고 함수내에서 알아서 작성되게 놔두면 되었음
