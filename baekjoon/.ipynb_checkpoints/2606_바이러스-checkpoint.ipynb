```python
import sys
from collections import deque

def bfs(i):
    global cnt
    visited[i] = 1
    q.append(i)
    while q:
        y = q.popleft()
        for x in range(comp + 1):
            if arr[y][x] and visited[x] == 0:
                q.append(x)
                visited[x] = 1
                cnt += 1


comp = int(sys.stdin.readline())
n = int(sys.stdin.readline())
temp = [list(map(int, sys.stdin.readline().split())) for _ in range(n)]
visited = [0] * (comp + 1)
q = deque()
arr = [[0] * (comp + 1) for _ in range(comp + 1)]
for i in temp:
    arr[i[0]][i[1]], arr[i[1]][i[0]] = 1, 1
cnt = 0
bfs(1)
print(cnt)
```
