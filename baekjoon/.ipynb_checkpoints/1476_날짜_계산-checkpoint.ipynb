```python
import sys

e, s, m = map(int, sys.stdin.readline().split())
N = [28 * 19, 15 * 19, 15 * 28]
dur = [13, 17, 10]
tot = e * N[0] * dur[0] + s * N[1] * dur[1] + m * N[2] * dur[2]
print((tot % 7980) if tot % 7980 else 7980)
```

---

#### 중국인의 나머지 정리

N의 배열은 **나머지 두 수의 곱**으로 구성<br>
dur 은 각 수의 역원(for 문으로 일일이 찾아본 값임<br>
tot 는 연립합동식의 해 구하는 과정
