```python
from math import gcd

n = int(input())
a = sorted([int(input()) for _ in range(n)])
cnt = []
g = a[1] - a[0]
for i in range(2, n):
    g = gcd(a[i] - a[i - 1], g)
ans = []
for i in range(1, int(g ** .5) + 1):
    if not g % i:
        ans.append(i)
        ans.append(g // i)
ans = sorted(list(set(ans)))
print(*ans[1:])
```
