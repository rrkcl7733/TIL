```python
import sys
input = sys.stdin.readline


def conq(x, y, n):
    global a, b, c
    tmp = [0, 0, 0]
    for i in range(x, x + n):
        tmp[0] += arr[i][y:y+n].count(-1)
        tmp[1] += arr[i][y:y+n].count(0)
        tmp[2] += arr[i][y:y+n].count(1)
    if tmp[0] == n ** 2:
        a += 1
    elif tmp[1] == n ** 2:
        b += 1
    elif tmp[2] == n ** 2:
        c += 1
    else:
        n //= 3
        for i in range(3):     # 9등분
            for j in range(3):
                conq(x + n * i, y + n * j, n)


n = int(input())
arr = [list(map(int, input().split())) for _ in range(n)]
a = b = c = 0
conq(0, 0, n)
print(a, b, c, sep='\n')
```
