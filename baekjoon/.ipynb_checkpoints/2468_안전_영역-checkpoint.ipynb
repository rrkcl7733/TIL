```python
import sys
from copy import deepcopy
sys.setrecursionlimit(10**6)

input = sys.stdin.readline


def dfs(i, j, cities):    #    0 
    cities[i][j] = 0
    for di, dj in [[-1, 0], [0, -1], [0, 1], [1, 0]]:
        ni, nj = i + di, j + dj
        if 0 <= ni < n and 0 <= nj < n and cities[ni][nj] != 0:
            dfs(ni, nj, cities)


def f(x):
    cnt = 0
    cities = deepcopy(arr)
    for i in range(n):    #     0 
        for j in range(n):
            if cities[i][j] <= x:
                cities[i][j] = 0
    for i in range(n):   # 0     0 
        for j in range(n):
            if cities[i][j] != 0:
                dfs(i, j, cities)
                cnt += 1
    return cnt


n = int(input())
arr = [list(map(int, input().rstrip().split())) for _ in range(n)]
maxV = 0
for i in range(n):      #   
    for j in range(n):
        if arr[i][j] > maxV:
            maxV = arr[i][j]

ans = []
for i in range(maxV):
    ans.append(f(i))     #    


print(max(ans))
```
