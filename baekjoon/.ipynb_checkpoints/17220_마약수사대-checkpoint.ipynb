```python
from collections import deque

pill = dict()
n, m = map(int, input().split())
for _ in range(m):
    a, b = input().split()
    pill[ord(a) - 65] = pill.get(ord(a) - 65, []) + [ord(b) - 65]
cut = input().split()[1:]
visited = [0] * 26
for i in cut:
    visited[ord(i) - 65] = 2

q = deque()
for k, v in pill.items():
    if not visited[k]:
        visited[k] = 1
        q.append(k)
    while q:
        x = q.popleft()
        for k, v in pill.items():
            if k == x:
                for j in v:
                    if not visited[j]:
                        q.append(j)
                        visited[j] = 1
print(visited.count(1) - 1)


# ord('A')-65
```
