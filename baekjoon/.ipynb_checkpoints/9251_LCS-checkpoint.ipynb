```python
a = input()
b = input()
n, m = len(a) + 1, len(b) + 1
LCS = [[0] * m for _ in range(n)]
for i in range(1, n):
    for j in range(1, m):
        if a[i - 1] == b[j - 1]:
            LCS[i][j] = LCS[i - 1][j - 1] + 1
        else:
            LCS[i][j] = max(LCS[i - 1][j], LCS[i][j - 1])
print(LCS[-1][-1])
```
