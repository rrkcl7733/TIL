## 메모리 초과 / 시간 초과

```python
import sys
input = sys.stdin.readline

def dfs(i, j, x):
    global maxV
    if i == j:
        if maxV < x:
            maxV = x
        return

    for di in range(1, n + 1):
        if tree[i][di] and not visited.get(di, 0):
            visited[di] = 1
            dfs(di, j, x + tree[i][di])
            visited[di] = 0


n = int(input())
tree = [[0] * (n + 1) for _ in range(n + 1)]
root = [0] * (n + 1)
for _ in range(n - 1):
    a, b, c = map(int, input().split())
    tree[a][b], tree[b][a] = c, c
    root[a] = 1    # 자식이 없는 노드 확인용

maxV = 0

for i in range(1, n):
    if not root[i]:    # 자식이 없는 노드들만
        for j in range(i + 1, n + 1):
            visited = dict()
            visited[i] = 1
            dfs(i, j, 0)

print(maxV)
```

---

#### 자식이 없는 노드들 중 답이 있다고 생각했고 그것들만 전부 dfs 로 구현하니 오류가 났다.

```python
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10**9)


def dfs(i, x):
    for di in tree[i]:
        s, cost = di
        if dist[s] == -1:
            dist[s] = x + cost
            dfs(s, dist[s])


n = int(input())
tree = [[] for _ in range(n + 1)]

for _ in range(n - 1):
    a, b, c = map(int, input().split())
    tree[a].append([b, c])
    tree[b].append([a, c])

dist = [-1] * (n + 1)
dist[1] = 0
dfs(1, 0)    # 루트 노드에서 거리 측정 1회

idx = dist.index(max(dist))    # 루트에서 가장 먼 놈 누구냐
dist = [-1] * (n + 1)
dist[idx] = 0
dfs(idx, 0)    # 거기서 거리 측정 1회

print(max(dist))
```

## 오류가 난 부분들<br>
- dist 거리 리스틩 초기 값을 0 미만의 값으로 설정해야했다. (0으로 하면 안됨) 근데 솔직히 왜 그런지는 모르겠음 가중치는 최소 1인데..<br>
- tree 설정을 10000 * 10000 이 아닌 들어오는 값마다 해당 노드에 append 하는 과정으로 메모리를 줄인다.<br>
