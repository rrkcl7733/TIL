```python
import sys
input = sys.stdin.readline

def dfs(i, j, x):
    global maxV
    if i == j:
        if maxV < x:
            maxV = x
        return

    for di in range(1, n + 1):
        if tree[i][di] and not visited.get(di, 0):
            visited[di] = 1
            dfs(di, j, x + tree[i][di])
            visited[di] = 0


n = int(input())
tree = [[0] * (n + 1) for _ in range(n + 1)]
root = [0] * (n + 1)
for _ in range(n - 1):
    a, b, c = map(int, input().split())
    tree[a][b], tree[b][a] = c, c
    root[a] = 1    #    

maxV = 0

for i in range(1, n):
    if not root[i]:    #   
        for j in range(i + 1, n + 1):
            visited = dict()
            visited[i] = 1
            dfs(i, j, 0)

print(maxV)
```

```python
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10**9)


def dfs(i, x):
    for di in tree[i]:
        s, cost = di
        if dist[s] == -1:
            dist[s] = x + cost
            dfs(s, dist[s])


n = int(input())
tree = [[] for _ in range(n + 1)]

for _ in range(n - 1):
    a, b, c = map(int, input().split())
    tree[a].append([b, c])
    tree[b].append([a, c])

dist = [-1] * (n + 1)
dist[1] = 0
dfs(1, 0)    #     1

idx = dist.index(max(dist))    #     
dist = [-1] * (n + 1)
dist[idx] = 0
dfs(idx, 0)    #    1

print(max(dist))
```
