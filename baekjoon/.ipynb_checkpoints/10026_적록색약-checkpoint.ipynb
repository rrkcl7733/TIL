```python
import sys
sys.setrecursionlimit(10**6)

def dfs(i, j):
    visited[i][j] = 1
    for di, dj in [[-1, 0], [0, -1], [0, 1], [1, 0]]:
        ni, nj = i + di, j + dj
        if 0 <= ni < n and 0 <= nj < n and not visited[ni][nj] and arr[i][j] == arr[ni][nj]:
            dfs(ni, nj)

n = int(input())
arr = [list(input()) for _ in range(n)]
visited = [[0] * n for _ in range(n)]
cnt_1 = cnt_2 = 0
for i in range(n):    #    dfs 
    for j in range(n):
        if not visited[i][j]:
            dfs(i, j)
            cnt_1 += 1

for i in range(n):    #  R, G   
    for j in range(n):
        if arr[i][j] == 'R':
            arr[i][j] = 'G'

visited = [[0] * n for _ in range(n)]    #  

for i in range(n):    # '' 
    for j in range(n):
        if not visited[i][j]:
            dfs(i, j)
            cnt_2 += 1

print(cnt_1, cnt_2)
```
