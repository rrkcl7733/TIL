```python
n, m = map(int, input().split())
arr = list(map(int, input().split()))
start, end = 1, max(arr)
while start <= end:
    point = (start + end) // 2
    tree = sum([i - point if i > point else 0 for i in arr])
    if tree >= m:
        start = point + 1
    else:
        end = point - 1
print(end)
```
