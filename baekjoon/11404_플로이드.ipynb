```python
import heapq
import sys
input = sys.stdin.readline
inf = int(1e9)

def dijkstra(start):
    q = []
    heapq.heappush(q, (0, start))    # (거리, 노드)
    distance[start] = 0
    while q:
        dist, now = heapq.heappop(q)
        if distance[now] < dist:     # 방문한 적 있었니?
            continue
        for i in graph[now]:
            cost = dist + i[1]     # 새로 발견한 길이 더 짧다면
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))


n = int(input())
e = int(input())
graph = [[] for _ in range(n + 1)]
for _ in range(e):
    a, b, c = map(int, input().split())
    graph[a].append((b, c))

for i in range(1, n + 1):
    distance = [inf] * (n + 1)
    dijkstra(i)
    distance = distance[1:]
    while distance.count(inf):
        distance[distance.index(inf)] = 0
    print(*distance)
```
