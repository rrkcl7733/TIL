```python
from copy import deepcopy
import sys
input = sys.stdin.readline


def f(arr):
    ans = 0
    while 1:
        new = [[-1] * n for _ in range(n)]
        nums = []
        group = []
        check = []
        for i in range(n):
            for j in range(n):
                q, people, temp, ssum = [], [], 0, 0
                q.append((i, j))
                while q:
                    x, y = q.pop()
                    if (x, y) in people:
                        continue
                    people.append((x, y))
                    for di, dj in [[-1, 0], [1, 0], [0, -1], [0, 1]]:
                        ni, nj = x + di, y + dj
                        if 0 <= ni < n and 0 <= nj < n and (ni, nj) not in check:
                            if a <= abs(arr[x][y] - arr[ni][nj]) <= b:
                                q.append((ni, nj))
                                check.append((ni, nj))
                                temp += 1
                                ssum += arr[ni][nj]
                if temp:
                    nums.append((temp, ssum))
                    group.append(people)
        for i in range(len(nums)):
            tot = nums[i][1] // nums[i][0]
            for x, y in group[i]:
                new[x][y] = tot
        for i in range(n):
            for j in range(n):
                if new[i][j] == -1:
                    new[i][j] = arr[i][j]
        arr = deepcopy(new)
        if not group:
            break
        ans += 1
    print(ans)


n, a, b = map(int, input().split())
arr = [list(map(int, input().split())) for _ in range(n)]
f(arr)
```
