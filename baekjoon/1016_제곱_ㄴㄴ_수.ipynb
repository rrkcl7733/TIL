#### 틀린 풀이

```python
m, n = map(int, input().split())
che = [0] * (n + 1)
for i in range(2, int(n ** .5) + 1):
    if che[i] != 1:
        x = i ** 2
        for j in range(x, n + 1, x):
            che[j] = 1

print(che[m:n + 1].count(0))
```

#### 맞는 풀이

```python
n, m = map(int, input().split())
che = [1] * (m - n + 1)
i = 2
while i ** 2 <= m:
    hubo = n // i ** 2
    while hubo * i ** 2 <= m:
        if n <= hubo * i ** 2 <= m:
            che[hubo * i ** 2 - n] = 0
        hubo += 1
    i += 1

print(sum(che))
```

**틀린풀이**의 이유 : 최대값입력한 수 크기의 배열 생성 후 `에라토스테네스의 체`의 방식으로 소수 판별 후 동시에 제곱수 들을 처리해주고 갯수 세려 했으나 입력값이 최대 1조.<br>
Python의 배열은 최대 3억의 크기까지 생성가능하므로 끝까지 배열생성하려면 실패.<br><br>
일일이 2부터 제곱수들의 배수들을 제거하기로 결정.<br>
배열의 크기는 min-max + 1 만큼.
