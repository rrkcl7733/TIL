```python
m, n = map(int, input().split())
che = [0] * (n + 1)
for i in range(2, int(n ** .5) + 1):
    if che[i] != 1:
        x = i ** 2
        for j in range(x, n + 1, x):
            che[j] = 1

print(che[m:n + 1].count(0))
```

```python
n, m = map(int, input().split())
che = [1] * (m - n + 1)
i = 2
while i ** 2 <= m:
    hubo = n // i ** 2
    while hubo * i ** 2 <= m:
        if n <= hubo * i ** 2 <= m:
            che[hubo * i ** 2 - n] = 0
        hubo += 1
    i += 1

print(sum(che))
```
