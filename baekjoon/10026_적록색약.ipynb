```python
import sys
sys.setrecursionlimit(10**6)

def dfs(i, j):
    visited[i][j] = 1
    for di, dj in [[-1, 0], [0, -1], [0, 1], [1, 0]]:
        ni, nj = i + di, j + dj
        if 0 <= ni < n and 0 <= nj < n and not visited[ni][nj] and arr[i][j] == arr[ni][nj]:
            dfs(ni, nj)

n = int(input())
arr = [list(input()) for _ in range(n)]
visited = [[0] * n for _ in range(n)]
cnt_1 = cnt_2 = 0
for i in range(n):    # 적록색약 아닌 사람 dfs 탐색
    for j in range(n):
        if not visited[i][j]:
            dfs(i, j)
            cnt_1 += 1

for i in range(n):    # 적록색약은 R, G 같은 취급 한대
    for j in range(n):
        if arr[i][j] == 'R':
            arr[i][j] = 'G'

visited = [[0] * n for _ in range(n)]    # 재설정 후

for i in range(n):    # '적록색약'인 경우
    for j in range(n):
        if not visited[i][j]:
            dfs(i, j)
            cnt_2 += 1

print(cnt_1, cnt_2)
```

적록색약인 경우 R, G를 같은 문자로 보는 것에 주목하여 한 가지 문자로 고쳐주는 작업을 거치고<br>
dfs 2회 반복.
