```python
import sys

n, k = map(int, sys.stdin.readline().split())
arr = [[0, 0]] + [list(map(int, sys.stdin.readline().split())) for _ in range(n)]
dp = [[0] * (k + 1) for _ in range(n + 1)]
for i in range(1, n + 1):
    for j in range(1, k + 1):
        w, v = arr[i]
        skip = dp[i - 1][j]
        if j < w:
            dp[i][j] = skip
        else:
            pick = dp[i - 1][j - w] + v
            dp[i][j] = max(pick, skip)
print(dp[n][k])
```
