```python
from collections import deque
import sys
from copy import deepcopy
input = sys.stdin.readline


def bfs():
    global maxV
    graph = deepcopy(arr)
    for i in range(n):
        for j in range(m):
            if arr[i][j] == 2:
               q.append((i, j))

    while q:
        i, j = q.popleft()
        for di, dj in [[-1, 0], [0, -1], [0, 1], [1, 0]]:
            ni, nj = i + di, j + dj
            if ni < 0 or nj < 0 or ni >= n or nj >= m:
                continue
            if not graph[ni][nj]:
                graph[ni][nj] = 2
                q.append((ni, nj))
    temp = 0
    for i in graph:
        temp += i.count(0)
    maxV = max(maxV, temp)


def make_wall(cnt):
    if cnt == 3:
        bfs()
        return
    for i in range(n):
        for j in range(m):
            if not arr[i][j]:
                arr[i][j] = 1
                make_wall(cnt + 1)
                arr[i][j] = 0


n, m = map(int, input().split())
arr = [list(map(int, input().rstrip().split())) for _ in range(n)]
q = deque()
maxV = 0
make_wall(0)
print(maxV)
```
