```python
a = int(input())
b = int(input())
x = int(input())
for i in range(a, b + 1):
    if x == sum(map(int, list(str(i)))):
        print(i)
        break
for i in range(b, a - 1, -1):
    if x == sum(map(int, list(str(i)))):
        print(i)
        exit()
```
