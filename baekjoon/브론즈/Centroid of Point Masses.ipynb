```python
i = 1
while 1:
    n = int(input())
    if n < 0:
        exit()
    x, y, m = 0, 0, 0
    for _ in range(n):
        a, b, c = map(int, input().split())
        x += a * c
        y += b * c
        m += c
    print(f'Case {i}: {x / m:.2f} {y / m:.2f}')
    i += 1
    input()
```
