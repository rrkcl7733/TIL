```python
for i in range(int(input())):
    print(f'Data Set {i + 1}:')
    n, w, e = map(int, input().split())
    x = 0
    for _ in range(n):
        a, b, c, d = map(int, input().split())
        x += max(w * a + e * c, w * b + e * d)
    print(x)
    print()
```
