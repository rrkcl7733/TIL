```python
for i in range(int(input())):
    a, b, n, r = map(int, input().split())
    z, c = 0, list(map(int, input().split()))
    for _ in range(n):
        x, y = map(int, input().split())
        if x in c:
            z += y * r // 100
    print(f'Data Set {i + 1}:')
    print(z)
    print()
```
