```python
a = b = c = 0
for i in range(int(input())):
    a1, b1, c1 = map(int, input().split())
    a += a1
    b += b1
    c += c1
    x = min(a, b, c)
    if x >= 30:
        print(x)
        a -= x
        b -= x
        c -= x
    else:
        print('NO')
```
