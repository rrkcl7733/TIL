```python
a1, b1, c1 = map(int, input().split(':'))
a2, b2, c2 = map(int, input().split(':'))
a2 -= a1
b2 -= b1
c2 -= c1
if c2 < 0:
    b2 -= 1
    c2 += 60
if b2 < 0:
    a2 -= 1
    b2 += 60
if a2 < 0:
    a2 += 24
if a2 > 23 or a2 == b2 == c2 == 0:
    a2, b2, c2 = '24', '00', '00'
a2, b2, c2 = str(a2), str(b2), str(c2)
a2 = '0' + a2 if len(a2) == 1 else a2
b2 = '0' + b2 if len(b2) == 1 else b2
c2 = '0' + c2 if len(c2) == 1 else c2
print(f'{a2}:{b2}:{c2}')
```
