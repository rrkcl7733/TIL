```python
n = int(input())
arr = []
for i in list(map(list, zip(*[list(map(int, input().split())) for _ in range(n)]))):
    b = [[] for _ in range(9)]
    for idx, x in enumerate(i):
        b[x - 1].append(idx + 1)
    for j in b:
        if len(j) > 1:
            arr.append(j)
ans = [set() for _ in range(n + 1)]
for i in arr:
    for j in i:
        for k in i:
            if j != k:
                ans[j].add(k)
a = list(map(len, ans))
a[0] = -1
print(a.index(max(a)))
```
