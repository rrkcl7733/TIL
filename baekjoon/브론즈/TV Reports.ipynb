```python
n, a, b = map(int, input().split())
for _ in range(n):
    x = int(input())
    if x > b:
        print(f'BBTV: Dollar reached {x} Oshloobs, A record!')
        b = x
    elif x < a:
        print(f'NTV: Dollar dropped by {a - x} Oshloobs')
    a = x
```
