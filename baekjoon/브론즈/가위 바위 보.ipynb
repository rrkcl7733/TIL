```python
for _ in range(int(input())):
    a = b = 0
    for _ in range(int(input())):
        x, y = input().split()
        if x == y:
            continue
        if (x, y) in [('R', 'S'), ('S', 'P'), ('P', 'R')]:
            a += 1
        else:
            b += 1
    print('Player 1' if a > b else 'Player 2' if a < b else 'TIE')
```
