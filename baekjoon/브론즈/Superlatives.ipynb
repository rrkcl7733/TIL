```python
for i in range(int(input())):
    a, b = map(int, input().split())
    z, cnt = a // b - 1, 0
    while z > 4:
        z //= 5
        cnt += 1
    c = 'no ' if a <= b else '' if z and not cnt else 'mega ' * cnt
    print(f'Data Set {i + 1}:')
    print(c + 'drought')
    print()
```
