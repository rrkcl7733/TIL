```python
ans = ('[', 0)
yeon = [0] * 4
team = [0] * 4
love = ['L', 'O', 'V', 'E']
x = input()
for i, j in enumerate(love):
    yeon[i] = x.count(j)
for _ in range(int(input())):
    x = input()
    for i, j in enumerate(love):
        team[i] = yeon[i] + x.count(j)
    tot = 1
    for i in range(4):
        for j in range(i + 1, 4):
            tot *= team[i] + team[j]
    tot %= 100
    if tot > ans[1]:
        ans = (x, tot)
    elif tot == ans[1]:
        ans = (sorted([x, ans[0]])[0], tot)

print(ans[0])
```
