```python
def check(i, j):
    for x in range(i, i + 3):
        for y in range(j, j + 3):
            if not a[x][y]:
                a[x][y] = 1
            else:
                a[x][y] = 0


n, m = map(int, input().split())
a = [list(map(int, list(input()))) for _ in range(n)]
b = [list(map(int, list(input()))) for _ in range(n)]
cnt = 0
if (n < 3 or m < 3 ) and a != b:
    print(-1)
    exit()
for i in range(n - 2):
    for j in range(m - 2):
        if a[i][j] != b[i][j]:
            cnt += 1
            check(i, j)
print(-1 if a != b else cnt)
```
