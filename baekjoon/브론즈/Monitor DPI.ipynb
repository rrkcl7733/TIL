```python
import decimal

context = decimal.getcontext()
context.rounding = decimal.ROUND_HALF_UP
while 1:
    d, h, v = map(float, input().split())
    if (d, h, v) == (0, 0, 0):
        exit()
    print(f'Horizontal DPI: {round(decimal.Decimal(str(h / (16 * d / 337 ** .5))), 2)}')
    print(f'Vertical DPI: {round(decimal.Decimal(str(v / (9 * d / 337 ** .5))), 2)}')
```
