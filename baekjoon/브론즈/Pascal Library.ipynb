```python
while 1:
    a, b = map(int, input().split())
    if not a + b:
        exit()
    x = list(map(int, input().split()))
    for _ in range(b - 1):
        x = [x + y for x, y in zip(x, list(map(int, input().split())))]
    print('yes' if x.count(b) else 'no')
```
