```python
n, m = map(int, input().split())
a = b = 0
arr = [input() for _ in range(n)]
arr2 = list(map(list, zip(*arr)))
for i in range(n):
    if 'X' not in arr[i]:
        a += 1
for i in range(m):
    if 'X' not in arr2[i]:
        b += 1
print(max(a, b))
```
