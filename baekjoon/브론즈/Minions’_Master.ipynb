```python
import decimal

context = decimal.getcontext()
context.rounding = decimal.ROUND_HALF_UP
for _ in range(int(input())):
    n, *a = map(int, input().split())
    x = round(decimal.Decimal(str(sum(a) / n)), 3)
    y = round(decimal.Decimal(str(sum([100 if i > x else 0 for i in a]) / n)), 3)
    print(f'{x} {y}%')
```

---

파이썬의 특성상 0.5의 반올림 값은 0이다.<br>그렇기 때문에 decimal의 초기 설정을 `ROUND_HALF_UP`으로 변경해주어 반올림 설정을 수정해야 한다.<br> 후 정확한 적용을 위해서 문자열으로 변경해야 한다.
