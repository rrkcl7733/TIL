```python
while 1:
    x = list(input().split())
    if x == ['#']:
        exit()
    a = b = 0
    for i in x:
        if i == '*':
            break
        elif i == 'A' or int(i) % 2:
            a += 1
        else:
            b += 1
    print('Cheryl' if a > b else 'Tania' if a < b else 'Draw')
```
