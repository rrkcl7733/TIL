```python
while 1:
    a = sorted(list(map(int, input().split())))
    if not sum(a):
        exit()
    print('Invalid' if a[2] >= a[0] + a[1] else 'Equilateral' if a[0] == a[1] == a[2] else 'Isosceles' if a[0] == a[1] or a[1] == a[2] else 'Scalene')
```
