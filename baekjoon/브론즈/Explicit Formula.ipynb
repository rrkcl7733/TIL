```python
a = list(map(int, input().split()))
x = a[0] or a[1]
for i in range(10):
    for j in range(i + 1, 10):
        if not i and j == 1:
            continue
        x ^= a[i] or a[j]
for i in range(10):
    for j in range(i + 1, 10):
        for k in range(j + 1, 10):
            x ^= a[i] or a[j] or a[k]
print(x)
```
